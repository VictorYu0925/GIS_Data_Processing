```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import geopandas as gpd
from matplotlib import cm
%matplotlib inline
```

# Draw Scatter on America

```python
plt.figure(figsize=(20,12))
map = Basemap(projection='stere',lat_0=90,lon_0=-105,
            llcrnrlat=23.41 ,urcrnrlat=45.44,
            llcrnrlon=-118.67,urcrnrlon=-64.52,
            rsphere=6371200.,resolution='l',area_thresh=10000)

# Draw boundary lines
map.drawmapboundary()
map.drawstates()
map.drawcoastlines()
map.drawcountries()

# Draw latitude, longitude lines
parallels = np.arange(0.,90,10.) 
map.drawparallels(parallels,labels=[1,0,0,0],fontsize=10)
meridians = np.arange(-110.,-60.,10.)
map.drawmeridians(meridians,labels=[0,0,0,1],fontsize=10)

# set Size by People
# people = np.array(case["People"],dtype=float)
# size=(pop/np.max(people))*1000
# x,y = map(lon,lat)
# map.scatter(x,y,s=size)

# Case, Control point
case = gpd.read_file('./geodata/usa/Case.gpx',encoding='utf-8')
x_p ,y_p = map(np.array(case['geometry'].x), np.array(case['geometry'].y))
map.scatter(x_p, y_p, c='#AA0000', label='Case')

control = gpd.read_file('./geodata/usa/Control.gpx',encoding='utf-8')
x_p ,y_p = map(np.array(control['geometry'].x), np.array(control['geometry'].y))
map.scatter(x_p, y_p, c='#0044BB', label='Control')

# print label
plt.legend(loc='lower right', prop={'size': 20})

plt.title('Case/Control Distribution', size = 20)
plt.show()
```

```python
# station data
stations = pd.read_csv('./geodata/usa/TMY3_StationsMeta.csv')
lat = np.array(stations["Latitude"])
lon = np.array(stations["Longitude"])
stations.head()
```

```python
plt.figure(figsize=(20,12))
map = Basemap(projection='stere',lat_0=90,lon_0=-105,
            llcrnrlat=23.41 ,urcrnrlat=45.44,
            llcrnrlon=-118.67,urcrnrlon=-64.52,
            rsphere=6371200.,resolution='l',area_thresh=10000)

# Draw boundary lines
map.drawmapboundary()
map.drawstates()
map.drawcoastlines()
map.drawcountries()

# Draw latitude, longitude lines
parallels = np.arange(0.,90,10.) 
map.drawparallels(parallels,labels=[1,0,0,0],fontsize=10)
meridians = np.arange(-110.,-60.,10.)
map.drawmeridians(meridians,labels=[0,0,0,1],fontsize=10)

# Draw Data
x,y = map(lon, lat)
map.scatter(x, y, c='#FF8800', label='TMY3 Weather Station')

# print label
plt.legend(loc='lower right', prop={'size': 20})

plt.title('TMY3 Weather Station Distribution', size = 20)
plt.show()
```
